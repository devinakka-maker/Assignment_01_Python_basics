**10.What are class and static methods in Python?**

- **Class Methods: @classmethod**
    - Accesses the class, not the instance.

```python
'''5.Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance.'''

class BankAccount:
  def __init__(self,bal):
    self.__balance=bal
  def deposit(self,b):
    self.__balance+=b
    return self.__balance
  def withdraw(self,b):
    if self.__balance >=b:
      self.__balance-=b
      return self.__balance
    else:
      return "insufficient balance"
  def check_balance(self):
    return self.__balance
obj=BankAccount(1000)
print(f"Balance is {obj.check_balance()}")
obj.deposit(2000)
print(f"Balance is {obj.check_balance()}")
obj.withdraw(500)
print(f"Balance is {obj.check_balance()}")



```

```python
'''7.Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers'''

class MathOperations:
  @classmethod
  def add_numbers(cls,x,y):
     cls.a=x
     cls.b=y
     print(f"addtion of two numbers {cls.a+cls.b}")
  @staticmethod
  def subtract_numbers(x,y):
    print(f"subtraction of two numbers {x-y}")

MathOperations.add_numbers(3,4)

MathOperations.subtract_numbers(4,2)
obj=MathOperations()
obj.subtract_numbers(4,2)




```

```python
'''10.Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors'''
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        raise TypeError("Operands must be of type Vector")

    def __str__(self):
        return f"({self.x}, {self.y})"


v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2

print(v3)
```

```python
'''14.Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary'''
class Employee:
  def __init__(self,h,hr):
    self.h=h
    self.hr=hr
  def calculate_salary(self):
    self.sal=self.h*self.hr
    return self.sal
    #print(f"salary is {sal}")
class Manager(Employee):
  def __init__(self,h,hr,b):
    super().__init__(h,hr)
    self.b=b
  def calculate_salary(self):
    self.sal=super().calculate_salary()
    self.sal+=self.b
    print(f"bonus is {self.sal}")
emp=Employee(5,1000)
print(f"salary is {emp.calculate_salary()}")
man=Manager(5,1000,5000)
man.calculate_salary()

```

```python
'''16.Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.'''

from abc import ABC,abstractmethod
class Animal:
  @abstractmethod
  def sound(self):
    pass
class Cow(Animal):
  def sound(self):
    print("Cow sound")

class Sheep(Animal):
  def sound(self):
    print("sheep sound")

obj=Cow()
obj.sound()

obj=Sheep()
obj.sound()


```
