**6.Explain the concept of generators in Python and how they are defined**
- A generator is a special type of function that yields values one at a time using the yield keyword instead of return.
- def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
- Features of Generators:
  - Lazy evaluation: Values are produced only when needed.
  - Memory efficient: No need to store the entire sequence in memory.
  - Stateful: Remembers where it left off between calls.
  - One-time use: Once exhausted, they can't be reused unless recreated.


**7.What are the advantages of using generators over regular functions?**

- Memory Efficiency
   - Generators produce values one at a time, using yield, rather than storing an entire sequence in memory.
   - Regular function: Returns all results at once (e.g., a list), consuming more memory.
   - Generator: Keeps only the current value in memory.
-  Lazy Evaluation
   - Generators compute values on demand, which means they don’t do any work until you ask for the next item.
   - Ideal for infinite sequences or delayed computations
   - functions return values at a time.

- State Retention
  - Generators remember their state between calls, making them perfect for tasks that require pause-and-resume behavior.

**9.Explain the purpose and usage of the `map()` function in Python.**
- The map() function in Python is a built-in tool that allows you to apply a function to every item in an iterable (like a list or tuple) and return a new iterable (a map object) with the results.
- Purpose of map()
  - Transforms data: Applies a function to each element of an iterable.
  - Avoids manual loops: Cleaner and often faster than using for loops.
  - Works with any iterable: Lists, tuples, sets, etc.
- syntax:
  - map(function, iterable)
- names = ["mike", "anu", "devi"]
capitalized = map(lambda x: x.capitalize(), names)
print(list(capitalized))  # Output: ['Mike', 'Anu', 'Devi']

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

   - Purpose: Applies a function cumulatively to reduce the iterable to a single value.
   - Returns: One final result.

   -from functools import reduce
    numbers = [1, 2, 3, 4]
    product = reduce(lambda x, y: x * y, numbers)

```python
'''5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.'''
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent
def power(n):
  #n=int(input("enter number"))
  l=[]
  for i in range(n+1):
    l.append(i)
  for i in l:
    yield i**2
p=power(4)
print(p.__next__())
```

```python
#7.Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filepath):
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print(f"File not found: {filepath}")
for line in read_file_lines('example.txt'):
    print(line)
```
