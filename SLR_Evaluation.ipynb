1.What is Simple Linear Regression (SLR)? Explain its purpose
- Simple Linear Regression (SLR) is one of the most fundamental techniques in statistics and machine learning. It models the relationship between two variables:
- Independent variable (X) - the predictor or input.
- Dependent variable (Y) - the outcome or target.
The relationship between X & Y is given by Y=m*X+c
- Purpose:
  -  Prediction of Y for given value X
  - Quantify how strongly x influences Y
  



2.What are the key assumptions of Simple Linear Regression?
- For simple linear regression to be valid, several assumptions must hold:
- Linearity: The relationship between X and Y is linear.
- Independence: Observations are independent of each other.
- Homoscedasticity: Constant variance of errors across values of X.
- Normality of residuals: Errors are normally distributed

3. Write the mathematical equation for a simple linear regression model and
explain each term.
- The mathematical equation of SLR is given by Y=m*X+c where
- Y (Dependent Variable / Response Variable)
  - The outcome we want to predict or explain.
- X (Independent Variable / Predictor Variable)
  - The input  we believe influences Y.
- c (Intercept)
  - The value of Y when X=0.
  - It represents the baseline level of the dependent variable.
- m(Slope / Regression Coefficient)
  - The amount by which Y changes when X increases by 1 unit.
  - It measures the strength and direction of the relationship.

5.What is the method of least squares in linear regression?
- The method of least squares is the most common technique used to estimate the parameters of a linear regression model. It finds the “best-fit” line by minimizing the total error between the predicted values and the actual observed values.
-  SLR is given by Ypred=m*X+c
The least squares method chooses c and m such that the sum of squared residuals is minimized





6.Name and briefly describe three common evaluation metrics for regression
models.
- Mean Absolute Error (MAE)
  - Definition:
The average of the absolute differences between predicted values and actual values.
  - Measures how far predictions are, on average, from the true values.
  - Pros: Easy to understand, less sensitive to outliers.
  - Cons: Doesn’t penalize large errors as strongly as squared metrics
-Mean Squared Error (MSE)
  - Definition:
The average of the squared differences between predicted and actual values.
  - Pros: Commonly used, mathematically convenient.
  - Cons: Sensitive to outliers.
- Rsquare:
  - Proportion of variance in the dependent variable explained by the model.
  - Rsq=explained variation/total variation
  - Interpretation:
Ranges from 0 to 1 (sometimes negative). Higher values mean better fit.
- Pros: Easy to interpret as “percentage of variance explained.

  




8.What is the purpose of the R-squared metric in regression analysis?
- The R-squared metric (also called the coefficient of determination) is used in regression analysis to measure how well the regression model explains the variability of the dependent variable.
-  Purpose of R-squared
   - Explains Variance
     - R-squared tells us the proportion of the variance in the dependent variable (Y) that is explained by the independent variable(s) (X).
- Model Fit Indicator
  - It provides a quick measure of how well the regression line fits the data.
  - Higher Rsq→ better fit (more variance explained).
  - Lower Rsq → poorer fit (less variance explained).
- Comparison Tool
  - Useful for comparing different regression models on the same dataset.
  - Helps decide which model explains the data better.


```python
'''9.Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept'''
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# X = independent variable (e.g., hours studied)
# y = dependent variable (e.g., exam scores)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # reshape for scikit-learn
y = np.array([2, 4, 5, 4, 5])

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Print slope (coefficient) and intercept
print("Slope :", model.coef_[0])
print("Intercept :", model.intercept_)
```
