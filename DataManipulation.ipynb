**1.What is the difference between multithreading and multiprocessing?**
Multithreading allows concurrent execution within a single process, while multiprocessing runs multiple processes in parallel, each with its own memory space.
- **Key differences:**
- **Multithreading**
   - Definition: Multiple threads run within a single process, sharing the same memory space.
   - Use Case: Ideal for I/O-bound tasks like file operations, network requests, or user input handling.
  - Performance: Limited by Python's Global Interpreter Lock (GIL), which allows only one thread to execute Python bytecode at a time.
  - Memory Usage: Efficient, since threads share memory.
  - Overhead: Lower overhead compared to multiprocessing.
  - Example Libraries: threading, concurrent.futures.ThreadPoolExecutor

- **Multiprocessing**
  - Definition: Multiple processes run independently, each with its own Python interpreter and memory space.
  - Use Case: Best for CPU-bound tasks like data processing, mathematical computations, or machine learning.
  - Performance: Not affected by the GIL, allowing true parallelism.
  - Memory Usage: Higher, as each process has its own memory.
  - Overhead: More overhead due to inter-process communication and memory duplication.
  - Example Libraries: multiprocessing, concurrent.futures.ProcessPoolExecutor




**2.What are the challenges associated with memory management in Python?**
- 1. Memory Leaks
  - Sometimes Python keeps data in memory even when it's no longer needed.
  - This happens if you forget to remove references to objects.

- 2. Garbage Collection
  - Python automatically cleans up unused memory.
  - But sometimes it runs too often or not enough, which can slow things down.

- 3. Circular References
  - If two objects refer to each other, Python may not know they’re no longer needed.
  - This can cause memory to stay used when it should be freed.

- 4. Fragmentation
  - Memory gets split into small pieces over time.
  - This makes it harder to use memory efficiently.

- 5. Big Objects Stay Too Long
  - Large lists or dictionaries might stay in memory longer than needed.
  - This can use up a lot of space

**6.What are the differences between NumPy arrays and Python lists**

NumPy arrays are faster, more memory-efficient, and better suited for numerical operations than Python lists, which are more flexible and can hold mixed data types.
- Performance & Speed
  - NumPy arrays are implemented in C and optimized for speed.
  - Python lists are slower for large-scale numerical computations.
  - Example: Adding two arrays is much faster with NumPy than looping through lists.

- Data Type Consistency
  - NumPy arrays are homogeneous — all elements must be of the same type (e.g., all floats).
  - Python lists are heterogeneous — can contain mixed types (e.g., [1, "hello", 3.5]).

- Memory Efficiency
  - NumPy arrays use less memory because they store data in contiguous blocks.
  - Python lists store references to objects, which takes more space.
- Mathematical operations
  - NumPy arrays supports element-wise operations directly
  - For lists we need to use the loops or list comprehensions.

  

7.Explain the difference between apply() and map() in Pandas

In Pandas, map() is used only with Series for element-wise operations, while apply() works with both Series and DataFrames and can apply functions to rows, columns, or entire structures.
- map() — For Series Only
  - Scope: Works only on a Pandas Series (one column).
  - Function Type: Applies a function to each element in the Series.
  - Use Case: Simple element-wise transformations.
- apply() — For Series and DataFrames
  - Scope: Works on both Series and DataFrames.
  - Function Type:
  - On Series: behaves like map() — applies function to each element.
  - On DataFrame: applies function to each row or column (you choose with axis).
  - Use Case: More flexible — can apply complex functions across rows or columns

```python
'''10.Calculate the correlation matrix using Seaborn and visualize it with a
heatmap.'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Math': [88, 92, 80, 89, 100],
    'Science': [90, 85, 88, 92, 95],
    'English': [70, 78, 85, 80, 75],
    'History': [85, 80, 78, 88, 90]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True,cmap='coolwarm')

plt.title("Correlation Matrix Heatmap")
plt.show()
```
