1.What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each

Artificial Intelligence (AI)
- Scope: The broadest concept — any technique that enables machines to simulate human intelligence.
- Techniques: Rule-based systems, expert systems, search algorithms, ML, DL, natural language processing.
- Applications: Chatbots, robotics, autonomous vehicles, medical diagnosis, game-playing (e.g., chess engines).
- Key Point: AI is the umbrella field under which ML and DL fall.
Machine Learning (ML)
- Scope: A subset of AI focused on algorithms that learn from data and improve over time.
- Techniques: Regression, decision trees, random forests, support vector machines, clustering.
- Applications: Spam filters, recommendation systems (Netflix, Amazon), fraud detection, predictive analytics.
- Key Point: ML is about learning patterns from data rather than hard-coded rules.
Deep Learning (DL)
- Scope: A specialized branch of ML using artificial neural networks with multiple layers (“deep” networks).
- Techniques: Convolutional Neural Networks (CNNs), Recurrent Neural Networks (RNNs), Transformers.
- Applications: Image recognition, speech-to-text, natural language understanding, self-driving cars.
- Key Point: DL excels at complex, high-dimensional data like images, audio, and text.
Data Science
- Scope: A multidisciplinary field combining statistics, programming, ML, and domain expertise to analyze and interpret data.
- Techniques: Data cleaning, exploratory data analysis, visualization, statistical modeling, ML integration.
- Applications: Business intelligence, customer analytics, healthcare predictions, financial modeling.
- Key Point: Data Science is about extracting insights and knowledge from data, often using AI/ML tools


2. Explain overfitting and underfitting in ML. How can you detect and prevent
them?

Overfitting
- Definition: Model fits training data extremely well but fails to generalize to new data.
- Symptoms:
- Very low training error, but high test/validation error.
- Complex models (deep trees, too many parameters) memorizing noise.
- Causes: Too much model complexity, insufficient training data, or lack of regularization.
- Prevention:
- Cross-validation: Evaluate model on multiple folds to ensure generalization.
- Regularization: Techniques like L1 (Lasso), L2 (Ridge), or dropout in neural networks penalize complexity.
- Early stopping: Stop training when validation error starts increasing.
- Simplify model: Reduce parameters or depth.
Underfitting
- Definition: Model is too simple and fails to capture underlying patterns.
- Symptoms:
- High error on both training and test sets.
- Predictions are overly simplistic.
- Causes: Model lacks complexity, poor feature selection, or insufficient training.
- Prevention:
- Use more complex models (e.g., deeper trees, more features).
- Improve feature engineering.
- Reduce regularization strength if it’s too restrictive.

Bias-Variance Tradeoff
- Bias: Error from overly simplistic assumptions (underfitting).
- Variance: Error from sensitivity to training data (overfitting).
- Goal: Find the sweet spot where both bias and variance are balanced for optimal generalization.


3.How would you handle missing values in a dataset? Explain at least three
methods with examples


following methods are used:
Deletion (Dropping Missing Values)
- Idea: Remove rows or columns that contain missing values.
- When to use:
- If the dataset is large and only a small percentage of values are missing.
- If the missing values are random and don’t bias the dataset.

Mean/Median Imputation
- Idea: Replace missing values with the mean or median of the column.
- When to use:
- For numerical data.
- Median is better if the data is skewed (e.g., salaries).

 Predictive Modeling (Advanced Imputation)
- Idea: Use machine learning models to predict missing values based on other features.
- When to use:
- When missing values are significant and you want more accurate imputation.


4.What is an imbalanced dataset? Describe two techniques to handle it

imbalanced datasets are a common challenge in machine learning, especially in classification problems like fraud detection,

Imbalanced Dataset:
- An imbalanced dataset occurs when the classes are not represented equally.
 Techniques to Handle Imbalance
1. Random Under/Oversampling
- Oversampling: Duplicate or generate more samples of the minority class to balance the dataset.
- Undersampling: Reduce the number of majority class samples to match the minority class.
- Pros: Simple to implement, improves balance quickly.
- Cons: Oversampling may cause overfitting (repeated data), undersampling may lose valuable information.
2. SMOTE (Synthetic Minority Oversampling Technique)
- Instead of duplicating minority samples, SMOTE generates synthetic samples by interpolating between existing minority class points.
- This creates more diverse and realistic samples, reducing overfitting.
- Pros: More robust than random oversampling, improves minority class representation.
- Cons: Can introduce noise if minority class is highly scattered.



5.Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.


feature scaling is a critical preprocessing step in machine learning. Let’s break it down clearly:

Feature Scaling is Important:
- Different ranges cause bias: Features like age (0–100) and salary (0–100,000) have vastly different scales. Algorithms may give undue importance to larger-scale features.
- Distance-based algorithms: Methods like KNN, SVM, and clustering rely on Euclidean distance. Without scaling, large-scale features dominate distance calculations.
- Gradient descent optimization: In models like linear regression, logistic regression, and neural networks, unscaled features cause gradients to oscillate, slowing convergence. Scaling ensures smoother and faster optimization
- Min-Max scaling is about range normalization (bounded values, good for distance-based methods).
- Standardization is about distribution normalization (centered data, good for models assuming Gaussian input).


6.Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Label Encoding and One-Hot Encoding are two common techniques for handling categorical variables in machine learning. Let’s compare them clearly:

Label Encoding
- What it does: Assigns each category a unique integer value.
Example:
Education = ["High School", "Bachelor", "Master", "PhD"]
Label Encoded → [0, 1, 2, 3]
- When to use:
- Best for ordinal categorical variables (where categories have a natural order, e.g., education level, satisfaction rating).
- Preserves the order relationship between categories.
- Limitations:
- If used on nominal variables (like colors or cities), the model may wrongly assume numeric relationships (e.g., "Red"=0, "Blue"=1, "Green"=2 → implies Green > Blue > Red, which is meaningless).

One-Hot Encoding
- What it does: Creates a new binary column for each category, marking presence with 1 and absence with 0.
Example:
Color = ["Red", "Blue", "Green"]
One-Hot Encoded →
Red   Blue   Green
1     0      0
0     1      0
0     0      1
- When to use:
- Best for nominal categorical variables (no inherent order, e.g., gender, city, color).
- Prevents the model from inferring false ordinal relationships.
- Limitations:
- Increases dimensionality (many new columns), which can be problematic with high-cardinality features (e.g., thousands of unique categories).
- Label Encoding → Use when categories are ordinal and order matters (e.g., education level, satisfaction rating).
- One-Hot Encoding → Use when categories are nominal and order does not matter (e.g., city names, product types).


```python
'''
7.Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?
'''
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("googleplaystore.csv")

# Drop rows with missing ratings
df = df.dropna(subset=["Rating"])

# Group by Category and calculate average rating
avg_ratings = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)

# Display top and bottom categories
print("Top 5 categories by average rating:")
print(avg_ratings.head())

print("\nBottom 5 categories by average rating:")
print(avg_ratings.tail())

# Plot average ratings per category
plt.figure(figsize=(12,6))
avg_ratings.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average App Ratings by Category")
plt.ylabel("Average Rating")
plt.xlabel("Category")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()




```

```python
'''
Question 8: Titanic Dataset
a) Compare the survival rates based on passenger class (Pclass). Which class had the highest
survival rate, and why do you think that happened?
'''
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")

survival_rates = df.groupby("Pclass")["Survived"].mean()

plt.bar(survival_rates.index, survival_rates.values, color=["green","orange","red"], edgecolor="black")
plt.title("Titanic Survival Rates by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.ylim(0,1)  
plt.show()

#Class 1 has the highest survival rate
```

```python
'''
Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?

'''
!pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("flight_price.csv")   # replace with your file name

# Quick check
print(df.head())

# Group by days_left and calculate average price
avg_price = df.groupby("days_left")["price"].mean().reset_index()

# Plot average price vs days left
plt.figure(figsize=(10,6))
sns.lineplot(x="days_left", y="price", data=avg_price, marker="o")
plt.title("Average Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left Until Departure")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

```python

import numpy as np
from scipy.optimize import curve_fit

# Define exponential function
def exp_func(x, a, b, c):
    return a * np.exp(b * x) + c

# Fit curve
x = avg_price["days_left"]
y = avg_price["price"]
params, _ = curve_fit(exp_func, x, y)

print("Exponential fit parameters:", params)
```

```python
'''
9b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
'''
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("flight_price.csv")
df

route_df = df[(df["source_city"] == "Delhi") & (df["destination_city"] == "Mumbai")]


avg_prices = route_df.groupby("airline")["price"].mean().sort_values()

print("Average prices for Delhi–Mumbai route:")
print(avg_prices)

# Plot bar chart
plt.figure(figsize=(10,6))
avg_prices.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Flight Prices by Airline (Delhi–Mumbai Route)")
plt.xlabel("Airline")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#AirAsia has the lowest rates
```
